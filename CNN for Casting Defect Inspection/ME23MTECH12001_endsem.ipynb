```python
# Ultrasonic Phased Array Defect Detection
# ME6040: Machine Learning and Its Applications

import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, callbacks
from tensorflow.keras.applications import MobileNetV2
from sklearn.model_selection import train_test_split
import cv2

# Mount Google Drive
from google.colab import drive
drive.mount('/drive')

# Define paths to the folders in Google Drive
images_dir = '/drive/MyDrive/Colab Notebooks/dataset_images'  # Path to images folder
labels_dir = '/drive/MyDrive/Colab Notebooks/dataset_label'   # Path to labels folder

# Check dataset structure
print(f"Number of images: {len(os.listdir(images_dir))}")
print(f"Number of labels: {len(os.listdir(labels_dir))}")

# Function to load and preprocess an image
def load_image(image_path, target_size=(224, 224)):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    img = cv2.resize(img, target_size)
    img = img / 255.0  # Normalize to [0,1]
    return img

# Function to parse YOLO format annotation
def parse_annotation(annotation_path, img_width=224, img_height=224):
    with open(annotation_path, 'r') as f:
        line = f.readline().strip().split()
        class_id = int(line[0])
        x_center = float(line[1]) * img_width
        y_center = float(line[2]) * img_height
        width = float(line[3]) * img_width
        height = float(line[4]) * img_height

        # Convert to [x_min, y_min, x_max, y_max] format
        x_min = int(x_center - width/2)
        y_min = int(y_center - height/2)
        x_max = int(x_center + width/2)
        y_max = int(y_center + height/2)

        return [x_min, y_min, x_max, y_max, class_id]

# Load dataset
image_paths = [os.path.join(images_dir, img_file) for img_file in os.listdir(images_dir)
               if img_file.endswith(('.jpg', '.jpeg', '.png'))]
label_paths = [os.path.join(labels_dir, os.path.splitext(os.path.basename(img_path))[0] + '.txt')
               for img_path in image_paths]

# Verify image and label pairs
valid_pairs = []
for img_path, label_path in zip(image_paths, label_paths):
    if os.path.exists(label_path):
        valid_pairs.append((img_path, label_path))
    else:
        print(f"Missing label for {img_path}")

print(f"Valid image-label pairs: {len(valid_pairs)}")

# Split data into training and validation sets
train_pairs, val_pairs = train_test_split(valid_pairs, test_size=0.2, random_state=42)
print(f"Training samples: {len(train_pairs)}, Validation samples: {len(val_pairs)}")

# Create custom dataset for object detection
class DefectDataset(tf.keras.utils.Sequence):
    def __init__(self, image_label_pairs, batch_size=16, img_size=(224, 224), augment=False):
        self.image_label_pairs = image_label_pairs
        self.batch_size = batch_size
        self.img_size = img_size
        self.augment = augment

    def __len__(self):
        return len(self.image_label_pairs) // self.batch_size

    def __getitem__(self, idx):
        batch_pairs = self.image_label_pairs[idx * self.batch_size:(idx + 1) * self.batch_size]

        batch_images = []
        batch_targets = []

        for img_path, label_path in batch_pairs:
            # Load and preprocess image
            img = load_image(img_path, self.img_size)

            # Parse annotation
            bbox = parse_annotation(label_path, self.img_size[0], self.img_size[1])

            # Create target mask (1 at defect location, 0 elsewhere)
            target = np.zeros((self.img_size[0], self.img_size[1], 1))

            # Draw the bounding box on the target
            x_min, y_min, x_max, y_max, _ = bbox
            target[y_min:y_max, x_min:x_max, 0] = 1.0

            batch_images.append(img)
            batch_targets.append(target)

        return np.array(batch_images), np.array(batch_targets)

# Create data generators
batch_size = 16
train_dataset = DefectDataset(train_pairs, batch_size=batch_size, augment=True)
val_dataset = DefectDataset(val_pairs, batch_size=batch_size)

# Visualize a few examples
def visualize_sample(dataset, index=0):
    images, masks = dataset[index]

    fig, axes = plt.subplots(3, 2, figsize=(12, 18))

    for i in range(3):
        # Display image
        axes[i, 0].imshow(images[i])
        axes[i, 0].set_title(f"Sample {i+1} - Image")
        axes[i, 0].axis('off')

        # Display mask
        axes[i, 1].imshow(masks[i].squeeze(), cmap='jet', alpha=0.7)
        axes[i, 1].set_title(f"Sample {i+1} - Defect Mask")
        axes[i, 1].axis('off')

    plt.tight_layout()
    plt.show()

# Uncomment to visualize training samples
# visualize_sample(train_dataset)

# Create a U-Net model for defect segmentation
def create_unet_model(input_shape=(224, 224, 3), output_channels=1):
    # Encoder (downsampling path)
    inputs = tf.keras.Input(shape=input_shape)

    # Encoder path
    conv1 = layers.Conv2D(64, 3, activation='relu', padding='same')(inputs)
    conv1 = layers.Conv2D(64, 3, activation='relu', padding='same')(conv1)
    pool1 = layers.MaxPooling2D(pool_size=(2, 2))(conv1)

    conv2 = layers.Conv2D(128, 3, activation='relu', padding='same')(pool1)
    conv2 = layers.Conv2D(128, 3, activation='relu', padding='same')(conv2)
    pool2 = layers.MaxPooling2D(pool_size=(2, 2))(conv2)

    conv3 = layers.Conv2D(256, 3, activation='relu', padding='same')(pool2)
    conv3 = layers.Conv2D(256, 3, activation='relu', padding='same')(conv3)
    pool3 = layers.MaxPooling2D(pool_size=(2, 2))(conv3)

    # Middle
    conv4 = layers.Conv2D(512, 3, activation='relu', padding='same')(pool3)
    conv4 = layers.Conv2D(512, 3, activation='relu', padding='same')(conv4)

    # Decoder path (upsampling)
    up5 = layers.UpSampling2D(size=(2, 2))(conv4)
    up5 = layers.Conv2D(256, 2, activation='relu', padding='same')(up5)
    concat5 = layers.Concatenate()([conv3, up5])
    conv5 = layers.Conv2D(256, 3, activation='relu', padding='same')(concat5)
    conv5 = layers.Conv2D(256, 3, activation='relu', padding='same')(conv5)

    up6 = layers.UpSampling2D(size=(2, 2))(conv5)
    up6 = layers.Conv2D(128, 2, activation='relu', padding='same')(up6)
    concat6 = layers.Concatenate()([conv2, up6])
    conv6 = layers.Conv2D(128, 3, activation='relu', padding='same')(concat6)
    conv6 = layers.Conv2D(128, 3, activation='relu', padding='same')(conv6)

    up7 = layers.UpSampling2D(size=(2, 2))(conv6)
    up7 = layers.Conv2D(64, 2, activation='relu', padding='same')(up7)
    concat7 = layers.Concatenate()([conv1, up7])
    conv7 = layers.Conv2D(64, 3, activation='relu', padding='same')(concat7)
    conv7 = layers.Conv2D(64, 3, activation='relu', padding='same')(conv7)

    # Output layer
    outputs = layers.Conv2D(output_channels, 1, activation='sigmoid')(conv7)

    # Create model
    model = models.Model(inputs=inputs, outputs=outputs)
    return model

# Alternative: Create a model using MobileNetV2 as backbone
def create_mobilenet_segmentation_model(input_shape=(224, 224, 3), output_channels=1):
    # Load MobileNetV2 as base model (without top layers)
    base_model = MobileNetV2(input_shape=input_shape, include_top=False, weights='imagenet')

    # Freeze the base model
    base_model.trainable = False

    # Get intermediate layers for skip connections
    skip_names = ['block_1_expand_relu', 'block_3_expand_relu', 'block_6_expand_relu', 'block_13_expand_relu']
    skips = [base_model.get_layer(name).output for name in skip_names]
    skips.reverse()

    # Get the output of the base model
    x = base_model.output

    # Upsampling path
    for i, skip in enumerate(skips):
        x = layers.UpSampling2D(size=(2, 2))(x)
        x = layers.Conv2D(256 // (2**i), 3, padding='same', activation='relu')(x)
        x = layers.Concatenate()([x, skip])
        x = layers.Conv2D(256 // (2**i), 3, padding='same', activation='relu')(x)

    # Final upsampling and output
    x = layers.UpSampling2D(size=(2, 2))(x)
    x = layers.Conv2D(32, 3, padding='same', activation='relu')(x)
    x = layers.UpSampling2D(size=(2, 2))(x)
    x = layers.Conv2D(16, 3, padding='same', activation='relu')(x)
    outputs = layers.Conv2D(output_channels, 1, activation='sigmoid')(x)

    # ADD THIS LINE HERE to resize from 448x448 to 224x224
    outputs = layers.Resizing(224, 224, interpolation='bilinear')(outputs)

    # Create model
    model = models.Model(inputs=base_model.input, outputs=outputs)
    return model


# Define IoU (Intersection over Union) metric
def iou_metric(y_true, y_pred):
    # Cast y_true to float32 to match y_pred
    y_true = tf.cast(y_true, tf.float32)

    intersection = tf.reduce_sum(y_true * y_pred, axis=[1, 2, 3])
    union = tf.reduce_sum(y_true, axis=[1, 2, 3]) + tf.reduce_sum(y_pred, axis=[1, 2, 3]) - intersection
    return tf.reduce_mean((intersection + 1e-7) / (union + 1e-7))

# Define Dice coefficient loss
def dice_loss(y_true, y_pred):
    numerator = 2 * tf.reduce_sum(y_true * y_pred, axis=[1, 2, 3])
    denominator = tf.reduce_sum(y_true + y_pred, axis=[1, 2, 3])
    return 1 - tf.reduce_mean((numerator + 1e-7) / (denominator + 1e-7))

# Initialize model (choose one)
# model = create_unet_model()
model = create_mobilenet_segmentation_model()

# Compile model
model.compile(
    optimizer=optimizers.Adam(learning_rate=1e-4),
    loss=dice_loss,
    metrics=[iou_metric, 'binary_accuracy']
)

# Model summary
model.summary()

# Define callbacks
callbacks_list = [
    callbacks.EarlyStopping(patience=10, restore_best_weights=True),
    callbacks.ReduceLROnPlateau(factor=0.1, patience=5),
    callbacks.ModelCheckpoint('/drive/MyDrive/Colab Notebooks/best_model.h5', save_best_only=True)
]

# Train model
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=2,
    callbacks=callbacks_list
)

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(history.history['iou_metric'], label='Training IoU')
plt.plot(history.history['val_iou_metric'], label='Validation IoU')
plt.title('IoU')
plt.xlabel('Epoch')
plt.ylabel('IoU')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(history.history['binary_accuracy'], label='Training Accuracy')
plt.plot(history.history['val_binary_accuracy'], label='Validation Accuracy')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

# Evaluate model on test data
val_loss, val_iou, val_acc = model.evaluate(val_dataset)
print(f"Validation Loss: {val_loss:.4f}")
print(f"Validation IoU: {val_iou:.4f}")
print(f"Validation Accuracy: {val_acc:.4f}")

# Function to visualize predictions
def visualize_predictions(dataset, model, num_samples=5):
    # Get a batch of samples
    images, true_masks = dataset[0]

    # Make predictions
    pred_masks = model.predict(images[:num_samples])

    # Visualize
    fig, axes = plt.subplots(num_samples, 3, figsize=(15, 4*num_samples))

    for i in range(num_samples):
        # Display original image
        axes[i, 0].imshow(images[i])
        axes[i, 0].set_title('Original Image')
        axes[i, 0].axis('off')

        # Display true mask
        axes[i, 1].imshow(images[i])
        axes[i, 1].imshow(true_masks[i].squeeze(), alpha=0.5, cmap='jet')
        axes[i, 1].set_title('True Defect')
        axes[i, 1].axis('off')

        # Display predicted mask
        axes[i, 2].imshow(images[i])
        axes[i, 2].imshow(pred_masks[i].squeeze(), alpha=0.5, cmap='jet')
        axes[i, 2].set_title('Predicted Defect')
        axes[i, 2].axis('off')

    plt.tight_layout()
    plt.show()

# Visualize predictions on validation data
visualize_predictions(val_dataset, model)

# Error analysis
def analyze_errors(dataset, model, threshold=0.5):
    total_samples = 0
    false_positives = 0
    false_negatives = 0

    for i in range(len(dataset)):
        images, true_masks = dataset[i]
        pred_masks = model.predict(images)

        # Apply threshold
        pred_binary = (pred_masks > threshold).astype(np.float32)

        for j in range(len(images)):
            total_samples += 1

            # Check if prediction has any positive pixels
            pred_has_defect = np.any(pred_binary[j] > 0)
            true_has_defect = np.any(true_masks[j] > 0)

            if pred_has_defect and not true_has_defect:
                false_positives += 1
            elif not pred_has_defect and true_has_defect:
                false_negatives += 1

    false_positive_rate = false_positives / total_samples
    false_negative_rate = false_negatives / total_samples

    print(f"Total samples analyzed: {total_samples}")
    print(f"False Positive Rate: {false_positive_rate:.4f}")
    print(f"False Negative Rate: {false_negative_rate:.4f}")

# Run error analysis
analyze_errors(val_dataset, model)

# Save model
model.save('/drive/MyDrive/Colab Notebooks/ultrasonic_defect_detection_model.h5')
print("Model saved successfully.")

# Function to convert predicted mask to bounding box
def mask_to_bbox(mask, threshold=0.5):
    # Apply threshold
    binary_mask = (mask > threshold).astype(np.uint8)

    # Find contours
    contours, _ = cv2.findContours(binary_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # If no contours found, return empty list
    if not contours:
        return []

    # Get bounding boxes for all contours
    bboxes = []
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        bboxes.append([x, y, x+w, y+h])

    return bboxes

# Experiment with different thresholds
def threshold_experiment(dataset, model, thresholds=[0.3, 0.5, 0.7, 0.9]):
    # Get a batch of samples
    images, true_masks = dataset[0]
    sample_image = images[0]
    sample_true_mask = true_masks[0]

    # Get prediction
    pred_mask = model.predict(np.expand_dims(sample_image, axis=0))[0]

    # Visualize with different thresholds
    fig, axes = plt.subplots(1, len(thresholds)+2, figsize=(15, 4))

    # Display original image
    axes[0].imshow(sample_image)
    axes[0].set_title('Original Image')
    axes[0].axis('off')

    # Display true mask
    axes[1].imshow(sample_image)
    axes[1].imshow(sample_true_mask.squeeze(), alpha=0.5, cmap='jet')
    axes[1].set_title('True Defect')
    axes[1].axis('off')

    # Display predictions with different thresholds
    for i, threshold in enumerate(thresholds):
        binary_pred = (pred_mask > threshold).astype(np.float32)

        # Display binary prediction
        axes[i+2].imshow(sample_image)
        axes[i+2].imshow(binary_pred.squeeze(), alpha=0.5, cmap='jet')
        axes[i+2].set_title(f'Threshold: {threshold}')
        axes[i+2].axis('off')

        # Get bounding box
        bboxes = mask_to_bbox(binary_pred.squeeze(), threshold=0.5)

        # Draw bounding box
        for bbox in bboxes:
            x_min, y_min, x_max, y_max = bbox
            rect = plt.Rectangle((x_min, y_min), x_max-x_min, y_max-y_min,
                               fill=False, edgecolor='red', linewidth=2)
            axes[i+2].add_patch(rect)

    plt.tight_layout()
    plt.show()

# Experiment with different thresholds
threshold_experiment(val_dataset, model)

# Final conclusions
print("Ultrasonic Phased Array Defect Detection Model Analysis:")
print("======================================================")
print(f"Final Validation IoU: {val_iou:.4f}")
print(f"Final Validation Accuracy: {val_acc:.4f}")
print("======================================================")
print("Model Architecture:")
print("- Used MobileNetV2 as backbone")
print("- Employed U-Net style skip connections")
print("- Optimized using Dice loss function")
print("======================================================")
print("Recommendations:")
print("1. For deployment, use a threshold of 0.5 for optimal balance")
print("2. Consider ensemble models for further performance improvements")
print("3. Collect more diverse defect samples for better generalization")
```
